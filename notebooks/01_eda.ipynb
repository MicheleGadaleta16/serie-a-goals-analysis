```python
# Import librerie
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Setup completo")
print(f"Pandas: {pd.__version__}")
print(f"Seaborn: {sns.__version__}")
```

```python
from pathlib import Path

# Path ai dati
DATA_DIR = Path("../data")

# Carica le tre stagioni
df_2223 = pd.read_csv(DATA_DIR / "serie_a_2022_23.csv")
df_2324 = pd.read_csv(DATA_DIR / "serie_a_2023_24.csv")
df_2425 = pd.read_csv(DATA_DIR / "serie_a_2024_25.csv")

# Unisci tutto in un unico DataFrame
df = pd.concat([df_2223, df_2324, df_2425], ignore_index=True)

print(f"Righe totali: {len(df)}")
print(f"Colonne: {df.columns.tolist()}")
```

```python
df_dropped=df.drop(['Div', 'Date', 'Time', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'B365H', 'B365D', 'B365A', 'BWH', 'BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'PSH', 'PSD', 'PSA', 'WHH', 'WHD', 'WHA', 'VCH', 'VCD', 'VCA', 'MaxH', 'MaxD', 'MaxA', 'AvgH', 'AvgD', 'AvgA', 'B365>2.5', 'B365<2.5', 'P>2.5', 'P<2.5', 'Max>2.5', 'Max<2.5', 'Avg>2.5', 'Avg<2.5', 'AHh', 'B365AHH', 'B365AHA', 'PAHH', 'PAHA', 'MaxAHH', 'MaxAHA', 'AvgAHH', 'AvgAHA', 'B365CH', 'B365CD', 'B365CA', 'BWCH', 'BWCD', 'BWCA', 'IWCH', 'IWCD', 'IWCA', 'PSCH', 'PSCD', 'PSCA', 'WHCH', 'WHCD', 'WHCA', 'VCCH', 'VCCD', 'VCCA', 'MaxCH', 'MaxCD', 'MaxCA', 'AvgCH', 'AvgCD', 'AvgCA', 'B365C>2.5', 'B365C<2.5', 'PC>2.5', 'PC<2.5', 'MaxC>2.5', 'MaxC<2.5', 'AvgC>2.5', 'AvgC<2.5', 'AHCh', 'B365CAHH', 'B365CAHA', 'PCAHH', 'PCAHA', 'MaxCAHH', 'MaxCAHA', 'AvgCAHH', 'AvgCAHA', 'BFH', 'BFD', 'BFA', '1XBH', '1XBD', '1XBA', 'BFEH', 'BFED', 'BFEA', 'BFE>2.5', 'BFE<2.5', 'BFEAHH', 'BFEAHA', 'BFCH', 'BFCD', 'BFCA', '1XBCH', '1XBCD', '1XBCA', 'BFECH', 'BFECD', 'BFECA', 'BFEC>2.5', 'BFEC<2.5', 'BFECAHH', 'BFECAHA'], axis=1)
```

```python
punteggio_macth = {'A': 0, 'D': 1, 'H': 3}
df_dropped['FTR']=df_dropped['FTR'].map(punteggio_macth)
print(df_dropped)
```

```python
h=df_dropped.groupby(["HomeTeam"]).sum()
a=df_dropped.groupby(["AwayTeam"]).sum()
gol_fatti_in_casa=pd.DataFrame(data=h, columns=['FTHG'])
gol_subiti_in_casa=pd.DataFrame(data=h, columns=['FTAG'])
gol_fatti_in_trasferta=pd.DataFrame(data=a, columns=['FTAG'])
gol_subiti_in_trasferta=pd.DataFrame(data=a, columns=['FTHG'])
```

```python
punti=pd.DataFrame(data=h, columns=['FTR'])
```

```python
punti.columns=['PUNTEGGIO']
```

```python
punti
```

```python
gol_fatti_in_casa.columns = ['gol']
gol_fatti_in_trasferta.columns = ['gol']
```

```python
gol_subiti_in_casa.columns = ['gol']
gol_subiti_in_trasferta.columns = ['gol']
gol_subiti = gol_subiti_in_casa + gol_subiti_in_trasferta
```

```python
gol_fatti=gol_fatti_in_casa+gol_fatti_in_trasferta
```

```python
gol_subiti
```

```python
df_finale = pd.concat([gol_fatti, gol_subiti], axis=1)
```

```python
df_finale.columns=['GOL FATTI', 'GOL SUBITI']
```

```python
df_finale
```

```python
df_finale_2= pd.concat([df_finale, punti], axis=1)
```

```python
df_finale_2
```

```python
df_finale_2.corr()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Grafico 1 — gol fatti vs punti

axes[0].scatter(df_finale_2['GOL FATTI'], df_finale_2['PUNTEGGIO'])
axes[0].set_xlabel('Gol fatti')
axes[0].set_ylabel('Punti')
axes[0].set_title('Gol fatti vs Punti (r=0.97)')



# Grafico 2 — gol subiti vs punti
axes[1].scatter(df_finale_2['GOL SUBITI'], df_finale_2['PUNTEGGIO'])
axes[1].set_xlabel('Gol subiti')
axes[1].set_ylabel('Punti')
axes[1].set_title('Gol subiti vs Punti (r=0.40)')

for squadra in df_finale_2.index:
    axes[0].annotate(
        text=squadra,
        xy=(df_finale_2.loc[squadra, 'GOL FATTI'], 
            df_finale_2.loc[squadra, 'PUNTEGGIO'])
    )

plt.tight_layout()
plt.savefig('../outputs/figures/correlation_goals_points.png', dpi=150)
plt.show()
```

```python
df_finale_2.sort_values(by='PUNTEGGIO', ascending=False)
```

```python
df_ordinato = df_finale_2.sort_values(by='PUNTEGGIO', ascending=True)

fig, ax = plt.subplots(figsize=(10, 12))
ax.barh(df_ordinato.index, df_ordinato['PUNTEGGIO'])
ax.set_xlabel('Punti')
ax.set_title('Classifica Serie A — 3 stagioni')
plt.tight_layout()
plt.savefig('../outputs/figures/classifica.png', dpi=150)
plt.show()
```
